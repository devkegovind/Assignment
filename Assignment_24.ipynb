**3. Compare and contrast map, filter, and reduce.**
- **reduce()** works differently than **map()** and **filter()**. **It does not return a new list based on the function and iterable we've passed. Instead, it returns a single value.** Also, in Python 3 reduce()isn't a built-in function anymore, and it can be found in the functools module.


###**The map() Function**
The `map()` function iterates through all items in the given iterable and executes the `function` we passed as an argument on each of them.

`map(function, iterable(s))`

```python
def starts_with_A(s):
  return s[0] == "A"

fruit = ["Apple", "Banana", "Pear", "Apricot", "Orange"]
map_object = map(starts_with_A, fruit)

print(list(map_object))
```

**By Using Lambda Function**

```python
fruit = ["Apple", "Banana", "Pear", "Apricot", "Orange"]
map_object = map(lambda s:s[0] == "A", fruit)
print(list(map_object))
```

```python
result_list = list(map(lambda s:s[0] == "A", fruit))

result_list
```

###**The filter() Function**
As the name suggest, `filter()` forms a new list that contains only elements that satisfy a certain condition, i.e. the `function` we passed returns `True`.

`filter(function, iterable(s))`

```python
# Without using lambdas

def starts_with_A(s):
  return s[0] == "A"

fruit = ["Apple", "Banana", "Pear", "Apriot", "Orange"]

filter_object = filter(starts_with_A, fruit)

print(list(filter_object))
```

```python
# By Using a lambda function

fruit = ["Apple", "Banana", "Pear", "Apriot", "Orange"]
filter_object = filter(lambda s:s[0] == "A", fruit)

print(list(filter_object))
```

```python
result_list1 = list(filter(starts_with_A, fruit))
result_list1
```

###**The reduce() Function**
`reduce()` works differently than `map()` and `filter()`. It does not return a new list based on the `function` and iterable we've passed. Instead, it returns a single value.
Also, in Python 3 `reduce()` isn't a buit-in function anymore, and it can be found in the `functools` module.
`reduce(fucntion, sequence[, initial])`

```python
from functools import reduce

def add(x, y):
  return x + y

list = [2,4,7,3]
print(reduce(add, list))
```

```python
# By Using lambda function

list = [2,4,7,3]
print(reduce(lambda x, y: x + y, list))
print("With an initial value: " + str(reduce(lambda x, y: x + y, list, 10)))
```
