**11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?**
- The python range() function returns the sequence of the given number between the given range.
- syntax:range(start, stop, step)
- **start**: integer starting from which the sequence of integer is to be returned.
- **stop**: integer before which the sequence of integers is to returned. The range of integer ends at stop -1.
- **step**: integer value which determines the increment between each integer in the sequence.
- range(stop)-- takes one argument.
- range(start, stop)-- takes two arguments.
- range(start, stop, step)-- takes three arguments.
- the difference between range(10), range(0, 10), and range(0, 10, 1) is range(10) contain only one argument, range(0,10) contain two arguments, range(0,10,1) contain three arguments.
- But output of all range function is same.
